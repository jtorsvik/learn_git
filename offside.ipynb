```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
def generate_coordinates(n:int) -> list:
    
    """
    This function creates a list of coordinates in a tuple.
    The number of coordinates in the list is determined by
    the input value n.

    The coordinates are set between x-range 0 and 50 and
    y-range between 0 and 100.

    input:
    n = Number of coordinates in the output list

    output:
    coordinates = List of n coordinates
    """

    x_range = (0, 50)
    y_range = (0, 100)

    coordinates = [(random.randint(x_range[0], x_range[1]), 
                   (random.randint(y_range[0], y_range[1]))) for _ in range(n)]

    return coordinates
```

```python
def generate_radius(n:int) -> list:

    """
    This function creates a list of radius distances.
    The number of instances is determined by the input
    value n.

    The radiuses are set at a maximum value equal to
    the hypotenuse of a triangle with side a == 50 and
    side b == 100.

    input:
    n = Number of radiuses in the output list

    output:
    coordinates = List of n radiuses
    """

    max_radius = np.sqrt(100**2 + 50**2)
    radius_range = (0, np.round(max_radius).astype(int))

    radius_list = [random.randint(radius_range[0], radius_range[1]) for _ in range(n)]

    return radius_list
```

```python
def generate_quarter_circle(radius, centre):

    angles = np.linspace(0, 2 * np.pi, 100)

    circle_x = centre[0] + radius * np.cos(angles)
    circle_y = centre[1] + radius * np.sin(angles)

    circle = pd.DataFrame({'x': circle_x,
                           'y': circle_y})
    
    # Conditions
    x_more_than_0 = (circle['x'] >= 0)
    y_more_than_0 = (circle['y'] >= 0)
    less_than_50 = (circle['x'] <= 50)
    less_than_100 = (circle['y'] <= 100)

    # quart = circle[x_more_than_0 & y_more_than_0]
    quart = circle[x_more_than_0 & y_more_than_0 & less_than_50 & less_than_100]

    return quart
```

```python
centre_1 = (0, 0)
centre_2 = (50, 0)
centre_3 = (0, 100)
centre_4 = (50, 100)
```

```python
quart1 = generate_quarter_circle(40, centre_1)
quart2 = generate_quarter_circle(40, centre_2)
quart3 = generate_quarter_circle(73.2, centre_3)
quart4 = generate_quarter_circle(73.2, centre_4)
```
